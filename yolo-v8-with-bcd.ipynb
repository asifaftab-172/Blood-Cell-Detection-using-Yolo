```python
from ultralytics import YOLO

model=YOLO("yolov8n.pt")
model.train(data="/kaggle/input/bcd-data/data.yaml", epochs=50)

```

```python
results = model.val(data='/kaggle/input/bcd-data/data.yaml',conf=0.2,iou=0.5,split='val')
results.confusion_matrix.matrix
```
